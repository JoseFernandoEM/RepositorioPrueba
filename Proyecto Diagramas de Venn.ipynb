```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

def main():
    elementos_a = input("Introduce los elementos del conjunto A separados por espacios: ").split()
    elementos_b = input("Introduce los elementos del conjunto B separados por espacios: ").split()
    
    conjunto_a = set(elementos_a)
    conjunto_b = set(elementos_b)

    venn2(subsets=(len(conjunto_a - conjunto_b), len(conjunto_b - conjunto_a), len(conjunto_a & conjunto_b)),
          set_labels=('Conjunto A', 'Conjunto B'))

    plt.title("Diagrama de Venn de dos conjuntos")
    plt.show()

    print("Elementos del conjunto A:", "{" + ", ".join(conjunto_a) + "}")
    print("Elementos del conjunto B:", "{" + ", ".join(conjunto_b) + "}")

    interseccion = conjunto_a & conjunto_b
    print("Los elementos en la intersección de los conjuntos son:", interseccion)

if __name__ == "__main__":
    main()
```

## 3 Conjuntos

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

def main():
    elementos_a = input("Introduce los elementos del conjunto A separados por espacios: ").split()
    elementos_b = input("Introduce los elementos del conjunto B separados por espacios: ").split()
    elementos_c = input("Introduce los elementos del conjunto C separados por espacios: ").split()

    conjunto_a = set(elementos_a)
    conjunto_b = set(elementos_b)
    conjunto_c = set(elementos_c)

    venn3(subsets=(len(conjunto_a - conjunto_b - conjunto_c), len(conjunto_b - conjunto_a - conjunto_c),
                   len(conjunto_a & conjunto_b - conjunto_c), len(conjunto_c - conjunto_a - conjunto_b),
                   len(conjunto_a & conjunto_c - conjunto_b), len(conjunto_b & conjunto_c - conjunto_a),
                   len(conjunto_a & conjunto_b & conjunto_c)),
          set_labels=('Conjunto A', 'Conjunto B', 'Conjunto C'))

    plt.title("Diagrama de Venn de tres conjuntos")
    plt.show()

    print("Elementos del conjunto A:", "{" + ", ".join(conjunto_a) + "}")
    print("Elementos del conjunto B:", "{" + ", ".join(conjunto_b) + "}")
    print("Elementos del conjunto C:", "{" + ", ".join(conjunto_c) + "}")

    interseccion_a_b = conjunto_a & conjunto_b - conjunto_c
    interseccion_a_c = conjunto_a & conjunto_c - conjunto_b
    interseccion_b_c = conjunto_b & conjunto_c - conjunto_a
    interseccion_a_b_c = conjunto_a & conjunto_b & conjunto_c
    print("Los elementos en la intersección de las regiones A ∩ B son:", interseccion_a_b)
    print("Los elementos en la intersección de las regiones A ∩ C son:", interseccion_a_c)
    print("Los elementos en la intersección de las regiones B ∩ C son:", interseccion_b_c)
    print("Los elementos en la intersección de las regiones A ∩ B ∩ C son:", interseccion_a_b_c)

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
from venn import venn

def ingresar_conjunto(nombre):
    datos = input(f"Ingrese los valores para {nombre} separados por comas: ")
    return set(map(int, datos.split(',')))

sets = {
    'Conjunto A': ingresar_conjunto('Conjunto A'),
    'Conjunto B': ingresar_conjunto('Conjunto B'),
    'Conjunto C': ingresar_conjunto('Conjunto C'),
    'Conjunto D': ingresar_conjunto('Conjunto D')
}

for nombre, conjunto in sets.items():
    print(f"{nombre}: {conjunto}")

print("\nIntersecciones entre los conjuntos:")
for nombre1, conjunto1 in sets.items():
    for nombre2, conjunto2 in sets.items():
        if nombre1 < nombre2:  
            interseccion = conjunto1 & conjunto2
            print(f"{nombre1} ∩ {nombre2}: {interseccion}")

venn(sets)

plt.show()
```
